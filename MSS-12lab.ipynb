```python
from math import sqrt

#==================================================================

#Функция расчёта геометрического числа Маха

#Вход - Критический диаметр, Диаметр
#Выход - Геометрическое значение числа Маха
def M(Dk,D): 
    m = 2
    
    for i in range(1000):
        m = 2.4/0.4 * pow( (D*D/(Dk*Dk)) * m, 0.8/2.4) - 2/0.4
        m = sqrt(m)

    return m
#==================================================================

M_glass = M(1.561, 1.725) #Значение геометрического числа Маха для стеклянного
M_alum = M(1.58, 2.272) #Значение геометрического числа Маха для алюминиевого

#==================================================================

#Функция определения нулевого давления по геометрическому числу Маха
def press0(M):
    k = pow((1+ (1.4-1)/2*M*M),-1.4/(1.4-1))
    p0 = 1/k
    return p0

#==================================================================

#Функция определения физического числа Маха

#Вход - Давление, Давление со штрихом, начальное число Маха (удобно брать геометрическое)
#Выход - Физическое значение числа Маха
def M2(p0,p0_st,m_start):
    k = 1.4
    m = m_start
    
    for i in range(10000):
        m = 0.4/2.8 + 2.4/2.8 * pow( (p0_st/p0), -0.4 ) * pow( 2 / 2.4 * (1/(m*m)+0.2), -1.4 )
        m = sqrt(m)

    return m

#==================================================================



p0_glass = press0(M_glass)
p0_alum = press0(M_alum)
print(p0_glass,p0_alum)
```
